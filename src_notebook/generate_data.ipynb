# Generate Data 

```python
import numpy as np # used for scientific computing\n",
import pandas as pd # used for data analysis and manipulation\n",
import matplotlib.pyplot as plt # used for visualization and plotting\n",
```

```python
num_rows = 1000
```

```python
def n_groups_data(N, d=0):
    latitude_lst = []
    longitude_lst = []
    for i in range(N):
        lat = latitude + (i * d)
        lon = longitude + (i * d)
        for _ in range(num_rows // N):
            dec_lat = np.random.random()/100
            dec_lon = np.random.random()/100

            latitude_lst.append(lat + dec_lat)
            longitude_lst.append(lon + dec_lon)
    
    return latitude_lst, longitude_lst
```

```python
# save and plot
Ns = [1, 3, 7, 9, 14, 21]
ds = [0.0005, 0.005, 0.05, 0.5]
fig, ax = plt.subplots(6, 4)
for i, N in enumerate(Ns):
    for j, d in enumerate(ds):
        latitude_lst, longitude_lst = n_groups_data(N, d)
        df = pd.DataFrame({'Latitude': latitude_lst, 'Longitude': longitude_lst})
        df.to_excel(f'data_N={N}_d={d}.xlsx', sheet_name='dataset')
        ax[i, j].scatter(df['Latitude'], df['Longitude'], color='blue')
        ax[i, j].set_title(f'Scatter - {N} group/s , d={d}')
fig.set_size_inches(22, 30)
plt.show()
```
